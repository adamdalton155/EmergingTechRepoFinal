<font color=blue><h3>Project Jupyter's History</h3>
    
JupyterLite isnt the first iteration of Project Jupyter so it makes sense to start where Jupyter started. Back in 2015, Project Jupyter was started with the goal to make an open source software for interactive computing across many programming languages. Its a branch of IPython in 2014 by Fernando PÃ©rez and Brian Granger and the name Jupyter comes from the three main languages it supports, that being Julia, Python and R. By 2015, more than 200,000 notebooks were available on GitHub and by January 2021, nearly 10 million notebooks were available. In 2021, Nature named Project Jupyter as one of the projects that transformed computer sciecnce
    
There have been four iterations of Project Jupyter so far, that being Jupyter Notebook, JupyterHub, JupyterLab and the latest JupyterLite. Since its development, many companies have adopted Jupyter, for example Google Colaboratory and Microsoft Azure Notebook to be used as a front end interface for cloud based users. Even Visual Studio Code supports local development of Jupyter notebooks. Jupyter Notebook is the first iteration was designed to be light weight and easy to use. It uses an ordered cell system of inputs and outputs which can either contain code or markdown language and underneath all of this is JSON (JavaScript Object Notation). Each file usually ends with the extension .ipynb. What makes Jupyter a pretty fast application is the fact that the user can choose which packages (e.g. numpy) they would like to import rather than having to load countless pre-installed packages and will run even faster if some of the statements have already been run as they may be strored in memory.

<font color=blue><h3>From Jupyter Notebook to Jupyter Lab</h3>
    
In early 2018, Version 1.0, the first stable release of Jupyter Lab released, the successor to Jupyter Notebook. The release of Lab was intended to make various improvements while keeping the lightweight, easy to use style of Notebook. I've listed some of the smaller improvements that Lab made below
    <ul>
        <li>Flexible Layout - The ability to customise the workspace. This includes drag and drop features, resize tabs and a collapsible sidebar that has the most used files easy to access.</li>
        <li>Reorganise cells - In Notebook, you had to add in cells either above or below by using 'A' or 'B' or by using the toolbar at the top. In Lab, you can drag and drop cells. Much easier!</li>
        <li>Switch to classic view - Some people are just going to prefer the old layout and luckily enough, there is a setting for the user to achieve this.</li>
        <li>Native support for Images, CSV, JSON, Markdown, PDF, Vega, Vega-Lite, etc. </li>
    </ul>
    
To me the biggest improvement in my opinion that Lab made over Notebook is the addition of terminal support. You can run Notebook through the terminal on the users machine and from there you can install packages but having that built into Lab greatly increases workflow. On this terminal, anything can be run including vim or emacs. Something else that I found interesting about the terminals on Lab is that the terminal actually runs where the Jupyter server itself is running.

<font color=blue><h3>JupyterLab and its short history</h3>

And now we come to the latest iteration of Project Jupyter, JupyterLab. Running entirely in the browser, and built from scratch while using components of JupyterLab and extensions, its being developed with the purpose of being lightweight and easy to access with a few clicks and all without having to install anything on the users machine. Even though JupyterLite is quite a bit away from a fully stable 1.0 release (0.1.0 beta version at the time of writing) its already benefitting from new features, accessibility fixes, and upkeep improvements because thats its built using the knowledge the developers have learned from developing JupyterLab. I have included two images below of the user interface just to show how similar Lite is to Lab.
    
JupyterLite ships natively with Pyolite, a Python kernel. The benefit of this for JupyterLab is now IPython is running in the browser, as so, features Code Completion (a feature I felt was missing while working on Jupyter Notebook), a nicer display for coding with and Interactive widgets. Also supported in JupyterLite is JavaScript and p5. Because of the purpose of Project Jupyter being designed mainly towards data science, having the ability to use JavaScript and p5 to code animations is a huge feature, helping data scientists to visualise their work in different ways. Like Pyolite, JavaScript and p5's kernels are supported but interestingly as an iFrame.
    
Deploying JupyterLite is actually quite simple. Its a static site that is served from a cacheable HTTP server, which in turn makes it so that notebooks from JupyterLite can be deployed from outside sources such as GitHub Pages, Binder and Vercel. Also, like JupyterNotebook, there is now a terminal command for launching JupyterLite. The command being 'jupyterlite'.
    
It is not perfect though. As mentioned above, JupyterLite is still in beta and a very early version at that. This does mean issues do occur. A quick google search and you can find Github repositories and discussions about packages working one minute and the next not. The cacheable nature of the project means that by accident, data loss can occur if say for example, the developers of your browser releases an update, that ends up clearing cache and as a result, work is deleted. On the other side of this, what if the user decides to clear their cache themselves in an attempt to fix an issue with the performance of their internet browser, not realising what would happen to their work. If a user worries about this situation occuring, maybe they should use JupyterLab. On Jupyter's blog, they did mention that eventually JupyterLab will completely replace JupyterNotebook so I believe it comes down to knowing what technology is appropriate for the users needs.

<font color=blue><h3>Examples of the packages capabilities</h3>

```python
#Numpy Example

import numpy as np
#Create an array 'a' of numbers that are mixed up
a = np.array([3,1,7,4,7])
#Sort the array from smallest to largest
np.sort(a)
```

```python
#Pandas Example

import pandas as pd
#Create data and within two strings (Apples and Oranges) with values for each
data = {
    'apples': [3, 2, 0, 1], 
    'oranges': [0, 3, 7, 2]
}
#Map the object data
purchases = pd.DataFrame(data)

#Print out purchases in a table view
purchases
```

```python
#Matplotlib Example

import matplotlib.pyplot as plt
#Plot the numbers for the graph
plt.plot([1, 3, 1.5, 4, 5, 8])
#Here we can label the X and Y axis with a label we can change
plt.ylabel('Y Values')
plt.xlabel('X Values')
#Print out the plotted graph
plt.show()
```

```python
#Seaborn Example

import seaborn as sns
#Sets the theme of the line plot
sns.set_theme(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
#Pillow Example

from PIL import Image
#Creating a new image that is 400px by 500px and set the Red, Green and Blue
img = Image.new("RGB",(400,500), (255, 0, 0))
#Generates the image and displays the image not on Jupyter but on the machine itself
img.show()
```
